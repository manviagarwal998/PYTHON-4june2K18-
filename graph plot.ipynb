```python
import matplotlib.pyplot as plt
x=[200,60,34,90]
y=["virat","rahul","sachin","yuvi"]
plt.xlabel("Time")
plt.ylabel("speed")
plt.bar(y,x,label="river")
#plt.bar(x1,y1,label="railway")
plt.legend()
plt.grid(color='y')
plt.show()
```
